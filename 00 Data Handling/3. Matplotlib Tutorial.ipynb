## Importing Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Line Plot

```python
x = [2, 5, 5, 6, 7, 8]
y = [3, 7, 7.2, 4, 12, 5]

plt.plot(x, y )
plt.plot(x, y, '^')
plt.title('Line Graph')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
```

### Scatter Plot

```python
x = [2, 5, 5, 6, 7, 8]
y = [3, 7, 7.2, 4, 12, 5]
     
plt.scatter(x, y)
plt.title('Line Graph')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.show
```

### Comparison

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]


plt.plot(x, y, color = 'g', linewidth = 6)
plt.plot(x2, y2, color = 'r', linewidth = 8)


plt.title("Line Plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.annotate(xy = [2, 3], text = 'Team_A')
plt.annotate(xy = [4, 4], text = 'Team_B')

plt.legend(['Team_A', 'Team_B'], loc = 2)

plt.show


# plt.savefig('foo.png')
```

## Vertical Bar Graph

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]

plt.bar(x, y, color = 'g', linewidth = 6, label = 'first')
plt.bar(x2, y2, color = 'r', linewidth = 4, label = 'second')

plt.title("Bar Graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')

# Adding LEGENDS
plt.legend(['First Line', 'Second Line'], loc=2)
plt.show
```

## Horizontal Bar Graph

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]

plt.barh(x, y, color = 'g', label = 'first')
plt.barh(x2, y2, color = 'r', label = 'second')

plt.title("Bar Graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')

# Adding LEGENDS
plt.legend(['First Line', 'Second Line'], loc=4)
plt.show
```

### Bubble Graph

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

plt.scatter(x, y, s = np.random.rand(5) * 1000, c = np.random.rand(5))

plt.title("Scater Plot")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show
```

## Pie Chart

```python
data = [3 ,7, 4, 12 ,5]
subject = ['a', 'b', 'c', 'd', 'e']
ex = [0, 0.2, 0, 0, 0]

plt.pie(data, labels = subject, explode = ex, shadow = True)
plt.show
```
